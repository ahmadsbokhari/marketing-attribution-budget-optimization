# Multi-Channel Marketing Attribution & Budget Optimization Model
Description: Developed a data-driven framework to analyze marketing channel performance and optimize a multi-million dollar annual budget.

Attribution Analysis: Implemented First-Touch Attribution logic to identify high-performing customer acquisition channels.

ROI Optimization: Calculated CPC (Cost Per Click), CAC (Customer Acquisition Cost), and ROI across diverse channels to identify inefficiencies.

Strategic Reallocation: Recommended a data-backed 20% budget increase strategy, achieving a 0.93 correlation between current ROI and recommended budget increases.

Insights: Identified key trends in "truck" vs. "sedan" sales, noting that 89.6% of truck sales lacked trade-ins, leading to a recommendation for targeted financing and trade-in incentive campaigns.

*Note: All data, channel names, and partner names in this notebook have been synthetically generated and anonymized. No real customer, spend, or sales data is used.*

### Part 1: Marketing Channel Efficiency & CPC Benchmarking

Search Engine has the lowest cost per click. The reason is that it is a pull marketing channel; users typically search with an intention to find a car, and this high intent makes them more likely to click the ad. In comparison, platforms like third-party listing and online video are push marketing channels with a higher CPC, because the ad must be compelling enough on its own to earn the click.

Despite having a similar CPC to social media, search engine is more profitable, with a slightly higher ROI and lower CAC, while also driving 46% of sales compared to just 11% from social media.

SearchPartnerA gives much bigger reach but a lower conversion rate.

### Part 5: Multi-Touch Customer Journey & Interaction Frequency Analysis

We had 96k users who made 169k clicks, 14k of which led to locks, leading to 1,348 sales.

Of the clicks that did not lead to a lock, only 35 users interacted with two channels, while the rest interacted with just one. The leading first-touch channels for this group were channel 2 (Search Engine-SearchPartnerA_Sedan), channel 14 (Social Media-SocialPlatformB), and channel 3 (Search Engine-SearchPartnerA_Truck).

Of the clicks that led to locks, around 600 locks had more than 20 clicks. This doesn't look like typical user behavior, and these may be bots, testers, researchers, or other outliers that skew the data significantly.

Excluding these users, we have 9,600 locks with an average of 6.3 clicks per lock, with 65% making 5 or fewer clicks. These users interacted with 4 channels on average, with a standard deviation of 2.74. The median is 3, meaning 50% of customers interact with 3 or fewer channels before locking a car. The most common channels for this group are channel 2 (Search Engine-SearchPartnerA_Sedan), channel 1 (Search Engine-SearchPartnerA_Convertible), and channel 5 (Search Engine-SearchPartnerB_Sedan).

It's interesting to note that only 27% of users locked on their first-touch channel, while 73% of locks came through a different channel, which highlights the importance of running multiple advertising platforms and building multiple marketing funnels. This is reinforced by the fact that, out of 235 distinct first-touch-to-lock pathways, the most common were conversions within search engine channels, with a large number of users starting with SearchPartnerA and locking with SearchPartnerB. The overwhelming majority of these users started with one bodystyle and locked on another, most commonly switching between Sedan and Convertible in either direction. This suggests that while users are aware of their need for a car, the type of vehicle they want can change once they see multiple options, a shift made easier through categories like search engine marketing.

### Part 6: 90-Day Conversion Lag & Non-Attributed Sales Analysis

For sales with click data available, we can define conversion in one of two ways: first-touch click to lock, and first-touch click to sale. Using the first definition, 327 sales (24.2%) have more than 90 days between first touch and lock. Using the second, 374 sales (27.7%) don't have a click within 90 days prior to the sale.

**First touch to lock:**
The median is 59 days, meaning 50% of confirmed sales convert within this window of consideration and locking. This also comes with a high standard deviation, and assuming a roughly normal distribution, about two-thirds of confirmed customers take between 24 and 94 days to lock. From a marketing perspective, campaigns should be designed to last within this span to capture most sales. Within this group, 46% come from the Search Engine category, with SearchPartnerA holding a 32% share of that. Finance partnerships make up 19% of these, which shows that their strength lies in converting long-deliberation customers.

Breaking this down further, for the 327 sales with more than 90 days in conversion, the only significant difference is in the make of car sold. The most frequently sold are Chevrolet, Mercedes-Benz, Hyundai, and Tesla, suggesting we can expect longer consideration times for these brands compared to others.

**First touch to sale:**
Consideration time is defined as the number of days between first touch and sale. A value greater than 90 days is categorized as high, and low otherwise.

Out of 1,348 sales, 40 (3%) have no click timestamp at all, which most likely means these customers came directly to the site and locked the vehicle. All of these took less than a month from lock to sale, but first-touch data isn't available. One possible explanation is that these users gathered information through sales partners or offline campaigns without a tracked interaction, then returned to complete the sale on their own. It's also possible these users weren't logged in or switched devices, so we don't have complete tracking data for them.

For the sales with click data available, 374 (27.7%) have a high consideration-time value.

17% of high values come from finance partnerships, which may reflect targeted efforts to clear inventory or outreach to customers on behalf of finance partners.

There is a positive change in high consideration time for makes like Toyota and Mercedes-Benz, suggesting customers take longer to decide on brands considered more reliable. Tesla also appears in the top five here, suggesting customers take longer to consider when switching to an EV, which is expected. For luxury brands like Audi, BMW, and Porsche, there is a negative change. Several factors could explain this, including brand loyalty, higher emotional purchase decisions, less research, or a younger audience, but more data is needed to confirm this. This suggests we can expect longer time-in-stock for the brands trending toward high consideration, and higher churn risk with a need for more aggressive campaigns for the brands trending toward low consideration. Marketing should focus on retargeting within the first two months for the latter group, or risk keeping that inventory in stock much longer than normal.

![Percentage Change from Low to High Consideration Time by Make](../images/part6_make_consideration_time.png)

For the high consideration-time group, the mean number of days between lock and sale is 4.6, with a high skew toward customers who take much longer to decide. The low consideration-time group has a mean of 2.74, with a much lower skew and about half the standard deviation, meaning these customers convert faster on average. Both groups have a median of 1 day, which is still good, considering 50% of these customers return once their mind is made up and don't deliberate much further.

There is no significant difference between the high and low consideration groups based on bodystyle, trade-in, financing, delivery distance, marketing channel, or category. There is a small difference based on marketing campaign, with 25+ having a 1% higher share than Convertible in the high consideration-time group, though this difference is not significant.

It's notable that even the low consideration-time group has a mean of 46 days between click and sale, compared to 107 days for the high consideration-time group. Marketing campaign timelines should account for this.

### Part 7: Data-Driven Budget Optimization & Growth Strategy

The chart below illustrates overall channel performance, plotting ROI against total spend.

![ROI vs. Spend by Channel](../images/part7_roi_vs_spend_scatter.png)

We ran a t-test (alpha = 0.05, critical value 1.96) to check whether increasing spend on a channel has a statistically significant correlation with number of sales. A t-value greater than 1.96, or a p-value less than 0.05, gives us 95% confidence that increasing the spend budget on that channel is likely to increase sales. The slope value tells us the expected increase in sales per additional dollar of spend.

Based on this analysis, FinancePartnerA is a clear winner. Looking at its sales data, 50% of its sales are financed, which is a major reason it drives profit up. I suggest a 2.5x increase in budget for FinancePartnerA, while keeping FinancePartnerB's budget unchanged (low t-value), and allocating $15k toward finding another partnership like FinancePartnerA.

SearchPartnerA is an excellent partner, particularly for the Sedan campaign. Its CPC and CAC are among the lowest, with a high ROI. However, all of its t-values are close to zero, meaning it may already be operating near its maximum efficiency, so we should continue running it as-is. I recommend increasing this budget by 20%, unless SearchPartnerA's costs rise.

SearchPartnerB_Sedan shows a positive trend even though its p-value is above 0.05. I suggest doubling the budget increase here, since all three metrics look favorable (ROI 3.23, p-value 0.17, and a positive slope).

Since many customers switch between bodystyles, we could also split the SearchPartnerB budget between Sedan and Convertible in a 3:2 ratio. This could be a good setup for future A/B testing.

In the social media category, I suggest increasing spend on SocialPlatformB by 60%. For SocialPlatformA, I suggest maintaining the current budget: while it delivers a good ROI, there's no clear evidence that more spend would increase sales. Seasonality or audience demographics may be limiting factors here.

Both third-party listing partners have an ROI below 1, meaning we're losing money on this channel. CAC and CPC are considerably high for both. Neither partner's customers financed their purchase, and only 2 of 124 traded in a vehicle. Average delivery distance (701 miles) was also higher than the overall average (627 miles). This suggests these customers are paying cash and are mostly located far from our facilities. ListingPartnerB has a high t-value, but we should still redesign these campaigns with financing and trade-in incentives, or cut them altogether. For this year, I suggest cutting the budget by 50% with redesigned campaigns, and cutting it entirely next year if performance doesn't improve.

Both Teen campaigns have extremely high CAC ($833 and $625) and an ROI below 1, so I suggest reducing spend here as well. Average delivery distance for these campaigns is 725 miles, compared to 625 miles overall, so I'd recommend redesigning these with geofencing, particularly for heavier payloads like SUVs (there were no truck sales in this segment). If we keep running these, they need better-performing creative, or we should cut them entirely.

Fortunately, truck sales have a lower average delivery distance (278 miles), so even though delivery cost eats further into margin, it isn't significantly high. However, 89.6% of truck sales had no trade-in, and 60.4% had no financing. I'd suggest the sales team consider this when pricing trucks, and marketing consider campaigns with trade-in and financing incentives.

This leaves 7% of the budget, which I suggest dividing equally (about $4,053 each) among the top 5 high-ROI campaigns, such as 25+, SearchPartnerA (excluding Sedan, since that budget was already increased by 20%), and SearchPartnerB Convertible.

The correlation between ROI and the recommended budget increase is 0.93, which confirms that we're increasing budget for the channels that already perform better. Channels to the left of the ROI = 1 line are losing money and are recommended for a budget decrease.
